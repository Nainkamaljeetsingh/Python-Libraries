```python
# Draw a radar chart

fig=plt.figure(figsize=(6,6))
ax=fig.add_axes([0.0,0.0,0.6,0.6],polar=True)

t=np.linspace(0,2*np.pi,100)

ax.plot(t,t,color='blue',lw=3)
```

```python
# Drawc a histogram

n=np.random.randn(100000)
fig,axes=plt.subplots(1,2,figsize=(12,4))
axes[0].set_title('Default histogram')
axes[0].hist(n)

axes[1].set_title('Cumulative detailed histogram ')
axes[1].hist(n,cumulative=True,bins=10)
```

```python
# Drawing a contour image

import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt


delta=0.025
x=np.arange(-3,3,delta)
y=np.arange(-2,2,delta)
X,Y=np.meshgrid(x,y)
Z1=np.exp(-X**2 - Y**2)
Z2=np.exp(-(X-1)**2 - (Y-1)**2)

Z=(Z1-Z2)*2 
```

```python
fig,ax=plt.subplots()
CS=ax.contour(X,Y,Z)
ax.clabel(CS,inline=1,fontsize=10)
```

```python
# Draw 3D  surface image

from mpl_toolkits.mplot3d.axes3d import Axes3D

fig=plt.figure(figsize=(14,6))
 

# specify the 3d graphics to draw with projection=3d

ax=fig.add_subplot(1,2,1,projection='3d')

ax.plot_surface(X,Y,Z,rstride=4,cstride=4,lw=0)
```

# Write  a Python  programming  to create   a pie chart  of the popularity  of the programming Languages.

```python
import matplotlib.pyplot as plt
# Data to plot

languages='python','java','php','javascript','c#','c++'

pop=[22.2,17.6,8.8,8,7.7,6.7]
colors=['#1f77b4','#ff7f0e','#2ca02c','#d62728','#9467bd','#8c564b'] 
```

```python
# explode 1st slice  

explode=(0.1,0,0,0,0,0)
plt.pie(pop,explode=explode,labels=languages,colors=colors,autopct='%1.1f%%',shadow=True,startangle=140)
```
